```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
pip install mlxtend
```

```python
df=pd.read_csv("Market_Basket_Optimisation.csv",header=None)
```

```python
df
```

```python
#converting them into list of list so that it can be handled by transaction encoder
records=[]
for i in range(len(df)):
    records.append([str(df.values[i,j]) for j in range(df.shape[1]) if pd.notna(df.values[i, j])])
records[0:10]
```

```python
len(records)
```

```python
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
x=te.fit_transform(records)
x
```

```python
len(te.columns_)
```

```python
df=pd.DataFrame(x,columns=te.columns_)
df.head()
```

```python
#finding frequent items and then findinh the rules
association_rules=apriori(records,min_support=0.01,min_confidence=0.25,min_lift=1.2,min_length=2)
association_results=list(association_rules)
```

```python
print(len(association_results))
```
